### Question 4
각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
# 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지의 channel 데이터를 channel_s로 저장
channel_s=channel[('2021-10-03 03'<channel['ct'])&(channel['ct']<='2021-11-01 15')]
```

```python
# channel_s를 channelname으로 그룹핑하여 subcnt의 처음, 마지막 값을 구해 channel_s_group으로 저장
channel_s_group=channel_s.groupby('channelname')['subcnt'].agg(['first','last'])
channel_s_group
```

```python
# last-first를 통해 subcnt의 증가량을 구해 del 컬럼으로 저장
channel_s_group['del']=channel_s_group['last']-channel_s_group['first']
channel_s_group
```

```python
# 결과와 같게 나오기 위해 index를 초기화하고 기존 인덱스인 channelname은 삭제하지 않음
# channelname과 del 출력
channel_s_group.reset_index(drop=False,inplace=True)
channel_s_group[['channelname','del']]
```
