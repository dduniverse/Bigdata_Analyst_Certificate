## 고객과 잘 맞는 타입 추천 :)
### basic1 데이터와 basic3 데이터를 'f4'값을 기준으로 병합하고, 
### 병합한 데이터에서 r2결측치를 제거한다음, 앞에서 부터 20개 데이터를 선택하고 'f2'컬럼 합을 구하시오

- basic1.csv: 고객 데이터 
- basic3.csv: 잘 어울리는 관계 데이터 (추천1:r1, 추천2:r2)

```python
import pandas as pd

df1=pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
df3=pd.read_csv('../input/bigdatacertificationkr/basic3.csv')
display(df1)
display(df3)
```

```python
# df1의 'f4' 컬럼 기준으로 df3과 병합
df=pd.merge(df1, df3, how='left', on='f4')
df
```

```python
# 'r2' 컬럼의 결측치 개수 확인
df['r2'].isnull().sum()
```

```python
# 'r2' 컬럼에 결측치가 있는 행을 삭제하고 결측치 개수 다시 확인
df.dropna(subset=['r2'],inplace=True)
df['r2'].isnull().sum()
```

```python
# 앞에서부터 20개의 데이터를 선택하여 'f2' 컬럼의 합계 구하기
df.head(20)['f2'].sum()
```

```python
# 라이브러리 및 데이터 로드
import pandas as pd
b1 = pd.read_csv("../input/bigdatacertificationkr/basic1.csv")
b3 = pd.read_csv("../input/bigdatacertificationkr/basic3.csv")
```

```python
b1.head()
```

```python
b3.head()
```

```python
# 데이터 결합(b1을 기준으로 결합) 
df = pd.merge(left = b1 , right = b3, how = "left", on = "f4")
df.head()
```

```python
df.tail()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# r2컬럼 결측치 제거
print(df.shape)
df = df.dropna(subset=['r2'])
print(df.shape)
```

```python
# 인덱스 리셋
df = df.reset_index()
```

```python
# 앞에서 부터 20개 데이터를 선택하고 'f2'컬럼 합
print(df.iloc[:20]['f2'].sum())
```
