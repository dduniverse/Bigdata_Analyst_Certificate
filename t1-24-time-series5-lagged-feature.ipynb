## 주어진 데이터(basic2.csv)에서 새로운 컬럼(1일 이전 시차 컬럼)을 만들고, Events가 1이면서 Sales가 1000000이하인 조건에 맞는 새로운 컬럼 합을 구하시오
- 데이터셋 : basic2.csv

```python
import pandas as pd

df=pd.read_csv('../input/bigdatacertificationkr/basic2.csv')
df
```

```python
# df 의 기본정보 확인
df.info()
```

```python
# Date 컬럼을 datetime 형식으로 변환
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
# 'PV' 컬럼을 아래로 한칸씩 내린 'p_PV' 컬럼 생성
df['p_PV']=df['PV'].shift(1)
df
```

```python
# 'p_PV' 컬럼의 결측치는 바로 다음 값으로 채움
df['p_PV']=df['p_PV'].fillna(method='bfill')
df
```

```python
# Events==1이고 Sales가 1000000이하인 p_PV의 합 구하기
df.loc[(df['Events']==1) & (df['Sales']<=1000000),'p_PV'].sum()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv")
df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
#1일 차이가 나는 시차 특성 만들기
df['previous_PV'] = df['PV'].shift(1)
df.head()
```

```python
# 1일 씩 미뤘음으로 가장 앞이 결측값이 됨 (바로 뒤의 값으로 채움)
df['previous_PV'] = df['previous_PV'].fillna(method = 'bfill')
df.head()
```

```python
# 조건에 맞는 1일 이전 PV의 합
cond = (df['Events'] == 1) & (df['Sales'] <= 1000000)
print(df[cond]['previous_PV'].sum())
```
